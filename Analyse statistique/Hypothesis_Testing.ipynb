# **Comparison of revenue's mean of offices in EMEA and those in NA**

```python
# Faire des jointures entre les tables et les rÃ©cuperer

df1 = pd.read_sql("Select o.* ," \
"e.*, c.*, p.* from offices o " \
"join employees e on e.officeCode = o.officeCode " \
"join customers c on c.salesRepEmployeeNumber = e.employeeNumber " \
"join payments p on p.customerNumber = c.customerNumber", con=engine)


df1
```

```python
df1.columns
```

```python
df1 = df1.loc[:, ~df1.columns.duplicated()]
```

```python
df2 = df1[df1["territory"].isin(["EMEA", "NA"])]
df2
```

```python
df3 = df2[["officeCode", "city", "state", "country", "employeeNumber",	"territory", "employeeNumber", "lastName",	"firstName",	"email",	"reportsTo", 	"jobTitle",
           	"customerNumber",	"customerName", "checkNumber",	"paymentDate",	"amount"]]
df3
```

## **Checking the normality of data**

```python
grp_EMEA = df3.loc[df3["territory"] == "EMEA"]
grp_EMEA
```

```python
grp_NA = df3.loc[df3["territory"] == "NA"]
grp_NA
```

```python
stats.shapiro(grp_EMEA["amount"])
```

Data are not normaly distributed, so let's apply non-parametric hypothesis testing

## **Applying the right test**

```python
stats.mannwhitneyu(grp_EMEA["amount"], grp_NA["amount"], alternative = "two-sided")
```

pvalue is closed to 1. That means, the two groups have almost the same revenue's mean

# **Analyzing whether sales are significantly different accross product lines or not**

```python
datf1 = pd.read_sql(
    """
    SELECT pdl.*, pd.*, od.*, o.*
    FROM productlines pdl
    JOIN products pd ON pdl.productLine = pd.productLine
    JOIN orderdetails od ON od.productCode = pd.productCode
    JOIN orders o ON o.orderNumber = od.orderNumber
    WHERE o.status = 'Shipped'
    """,
    con=engine
)

datf1
```

```python
datf1 = datf1.loc[:, ~datf1.columns.duplicated()]
datf1
```

```python
datf1.columns
```

```python
datf2 = datf1[['productLine',
       'productCode', 'productName', 'productScale', 'buyPrice', 'MSRP',
       'orderNumber', 'quantityOrdered', 'priceEach']]
datf2
```

```python
datf2["Amount"] = datf2["priceEach"]*datf2["quantityOrdered"]
```

```python
datf2["productLine"].unique()
```

```python
grp_vint = datf2[datf2["productLine"] == 'Vintage Cars']
grp_clas = datf2[datf2["productLine"] == 'Classic Cars']
grp_truc = datf2[datf2["productLine"] == 'Trucks and Buses']
grp_tr = datf2[datf2["productLine"] == 'Trains']
grp_ships = datf2[datf2["productLine"] == 'Ships']
grp_planes = datf2[datf2["productLine"] == 'Planes']
grp_moto = datf2[datf2["productLine"] == 'Motorcycles']
```

## **Checking the normality of data**

```python
stats.shapiro(grp_vint["Amount"])
```

Data are not normaly distributed, so let's apply non-parametric hypothesis testing

## **Applying the right test**

```python
stats.kruskal(grp_vint["Amount"], grp_clas["Amount"], grp_truc["Amount"], 
              grp_tr["Amount"], grp_ships["Amount"], grp_planes["Amount"], grp_moto["Amount"])
```

pvalue is closed to 0. That means sales are significanty diffrenent accross productlines
